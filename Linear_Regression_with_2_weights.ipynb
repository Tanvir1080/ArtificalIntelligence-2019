```
%matplotlib inline
from mpl_toolkits import mplot3d
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```
# Get our test data 
np.random.seed(42)

totalValues = 100
xMax = 2
x1Values = xMax * np.random.rand(totalValues, 1)
x2Values = xMax * np.random.rand(totalValues, 1)

groundTruth = 8
weight1 = 7
weight2 = 4

yValues = groundTruth + (weight1 * x1Values) + (weight2 * x2Values) + np.random.rand(totalValues, 1)
```

```
# Plot what we have but in 3D cuz 2 Xs

fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')
ax.set_xlabel('X1')
ax.set_xlabel('X2')
ax.set_ylabel('Y')
scatter = ax.scatter3D(x1Values, x2Values, yValues)
```

```
# Split data into training and testing sets 
# Training will be 80% of data and 20% for testing
train = int(len(x1Values) * 0.80)
test = totalValues - train

trainX1 = x1Values[:train]
trainX2 = x2Values[:train]
trainY = yValues[:train]

testX1 = x1Values[train:]
testX2 = x2Values[train:]
testY = yValues[train:]
```

# Linear regression using numpy


```
# number of epochs
epochs = 10

# learning rate
lr = 0.01

w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)
```

```
for epoch in np.arange(epochs):
  for i in np.arange(train):
    y_pred = w1 * trainX1[i] + w2 * trainX2[i] + b
    
    grad_w1 = (y_pred - trainY[i]) * trainX1[i]
    grad_w2 = (y_pred - trainY[i]) * trainX2[i]
    grad_b = (y_pred - trainY[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b -= lr * grad_b
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * ((w1 * testX1[i] + w2 * testX2[i] + b) - testY[i]) ** 2
test_loss /= 20
```

```
test_loss
```

```
predY = w1 * testX1 + w2 * testX2 + b
```

```
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
ax.set_xlabel('X1')
ax.set_xlabel('X2')
ax.set_ylabel('Y')
scatter = ax.scatter3D(testX1, testX2, testY, color='green')
scatter = ax.scatter3D(testX1, testX2, predY, color='red')
```
