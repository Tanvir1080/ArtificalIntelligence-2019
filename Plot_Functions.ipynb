```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
# First function, local maximum
# Generating a reverse parabola 
def sumSquares(x, y): 
  return -(x**2) - (y**2)

x = np.linspace(-100, 100)
y = np.linspace(-100, 100)

X, Y = np.meshgrid(x, y)
z = sumSquares(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot([0], [0], [0], marker='.', markersize=15, alpha=1, zorder=100)
ax.text(15, 0.5, 1500,'Local Maxima (0,0,0)', zorder=100)
ax.contour3D(X, Y, z, 50, cmap='binary')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z');
```

```
# Second function, local minimum
# Generating a regular parabola 
def sumSquares(x, y): 
  return (x**2) + (y**2)

x = np.linspace(-100, 100)
y = np.linspace(-100, 100)

X, Y = np.meshgrid(x, y)
z = sumSquares(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot([0], [0], [0], marker='.', markersize=15, alpha=1, zorder=100)
ax.text(15, 0.5, 1500,'Local Maxima (0,0,0)', zorder=100)
ax.contour3D(X, Y, z, 50, cmap='binary')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z');
```

```
# Third function, gradient is 0 but not at a local min or max 
# This graph generates a saddle point graph

def sumSquares(x, y): 
  return (x**2) - (y**2)

x = np.linspace(-100, 100)
y = np.linspace(-100, 100)

X, Y = np.meshgrid(x, y)
z = sumSquares(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot([0], [0], [0], marker='.', markersize=15, alpha=1, zorder=100)
ax.text(15, 0.5, 1500,'Local Maxima (0,0,0)', zorder=100)
ax.contour3D(X, Y, z, 50, cmap='binary')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z');
```
